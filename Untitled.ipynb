```python
import matplotlib.pyplot as plt
import numpy as np
#from SyMBac.PSF import PSF_generator
import tifffile
```

```python
radius = 30
pix_mic_conv = 0.11*60/100
apo_sigma = 0
NA = 0.95
n = 1
wavelength = 0.55
resize_amount = 1
FL_PSF = PSF_generator(
    radius = radius,
    wavelength = wavelength,
    NA = NA,
    n = n,
    resize_amount = resize_amount,
    pix_mic_conv = pix_mic_conv,
    apo_sigma = apo_sigma,
    mode="3d fluo",
    condenser = "Ph3",
    z_height = int(np.ceil(2/pix_mic_conv)),
    pz = 0.5,
    working_distance = 89
)

FL_PSF.calculate_PSF()

FL_PSF.plot_PSF()
tifffile.imwrite("60x_NA_0.95_n_1_wavelength_550.tiff", FL_PSF.kernel)
```
